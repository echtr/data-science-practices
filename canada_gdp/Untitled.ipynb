```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("canada_gdp.csv")
```

```python
df
```

```python
df.shape, df.isnull().sum(), df.dtypes
```

```python
df["income"] = df["income"].astype("int")
df.dtypes
```

```python
df
```

```python
x = df["year"].to_numpy()
y = df["income"].to_numpy()
len(x) == len(y)
```

```python
plt.figure(figsize=(24, 9))
plt.scatter(x, y)
plt.xlabel("years")
plt.ylabel("income")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 5/47, random_state=1)
```

```python
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
```

```python
reg = LinearRegression().fit(x_train, y_train)
```

```python
reg.predict(x_test)
```

```python
y_test
```

```python
reg.score(x_test, y_test)
```

```python
px = reg.predict(x.reshape(-1,1)) #predict x
plt.figure(figsize=(24, 9))
plt.scatter(x, y)
plt.plot(x, px, color="red")
plt.xlabel("years")
plt.ylabel("income")
plt.show()
```
